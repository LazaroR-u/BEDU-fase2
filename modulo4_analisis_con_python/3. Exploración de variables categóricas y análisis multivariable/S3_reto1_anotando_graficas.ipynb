##Visualización de diagnósticos de mujeres con o sin diabetes de la comunidad indigena Pima.

```python
df = pd.read_csv('/content/drive/MyDrive/BEDU_DS/FASE_3/Remoto Datasets/Remoto diabetes-clean.csv', index_col=0)

df.head()
```

### a) Agregamos un titulo al histograma de glucosa de la poblacion de indigenas de Pima.

```python
ax=sns.histplot(df['glucose'], kde=False)
ax.set(title="histograma de glucosa de indigenas pima", xlabel="glucosa", ylabel="conteo");
```

### b) histograma de numero de embarazos por mujer de la comunidad indigena Pima

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
br = sns.barplot(pregnancy_counts.index, pregnancy_counts)
br.set(title="numero de embarazos de mujeres de india Pima", xlabel="numeros de embarazo", ylabel="conteo")
```

###Comparación de la densidad de frecuencia de los niveles de glucosa entre personas con diabetes y sin diabetes de la comunidad Pima

```python
sns.displot(data=df, x='glucose', hue='outcome', kind='kde', common_norm=False, fill=True)
```

### c) Densidad de frecuencia comparativa de los niveles de glucosa entre las personas con diabetes y las personas sin diabetes. 

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
```

### d) Densidades de frecuencia de la presion sanguinea, niveles de glucosa e insulina de las personas con diabetes.

```python
sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True})
```
